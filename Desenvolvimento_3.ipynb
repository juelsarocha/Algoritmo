```python
#Imprimir todos os número exceto o 13 (laço 'for... in ranger)
for i in range(1, 21) :
  if(i ==1) :
    continue
  else :
    print(1)



#Imprimir todos os números exceto o 13 (laço 'while')
contador = 1
while (contador <=20) :
  if(contador ==13) :
    contador = contador + 1
    continue
  else :
    print(contador)
    contador = contador + 1


#Imprimir todos os números exceto o 13 em ordem descendente
for i in range(20, 0 -1) :
  if(i == 13) :
    continue
  else:
    print(i)
```
